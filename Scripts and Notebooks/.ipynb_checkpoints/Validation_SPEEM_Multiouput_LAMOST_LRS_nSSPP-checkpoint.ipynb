```python
import pickle

tiles, acor = pickle.load(open("jplus_dr2_6total.pkl", "rb"))

Aper_corrs = {'TILE_ID' : tiles, 'u_ApCor' : acor[:,0], 'J0378_ApCor' : acor[:,1],
              'J0395_ApCor' : acor[:,2], 'J0410_ApCor' : acor[:,3], 'J0430_ApCor' : acor[:,4],
              'g_ApCor' : acor[:,5], 'J0515_ApCor' : acor[:,6], 'r_ApCor' : acor[:,7],
              'J0660_ApCor' : acor[:,8], 'i_ApCor' : acor[:,9], 'J0861_ApCor' : acor[:,10],
              'z_ApCor' : acor[:,11]}

acor_6 = pd.DataFrame(Aper_corrs)
```

```python
stellar = ['INSPECT', 'SNR', 'SPECTYPE_SUBCLASS', 'Teff_GAIA',
           'TEFF_ADOP', 'LOGG_ADOP', 'FEH_ADOP', 'FEH_ANNRR',
           'TEFF_SPEC', 'LOGG_SPEC', 'FEH_SPEC', 'TEFF_ADOP_UNC',
           'FEH_ADOP_UNC', 'LOGG_ADOP_UNC']


Filters = list(AstroTools.Filters['JPLUS'])

def correctMags(df):
    
    Filters = list(AstroTools.Filters['JPLUS'])
    Correction = list(AstroTools.Filters['Ax'])
    A_Correction = list(acor_6.columns[1:])
    
    dfcp = pd.concat([df_result[a].sub(df_result[b]).add(df_result[c]) for a, b, c in zip(Filters,Correction, A_Correction)],axis=1, keys=Filters).round(3)
    
    return dfcp

    
    return dfcp
    
def add_id_columns(df1, df2, columns=AstroTools.Reference['Full']):
    
    df = df1.join(df2[columns])
    
    return df

def order_columns(df,reference=AstroTools.Reference['Full'],
                  filters=AstroTools.Filters['JPLUS']):
    
    return df[reference + filters]

def add_gaia_filters(df1, df2, columns=AstroTools.Filters['GAIA']):
    
    df = df1.join(df2[columns])
    
    return df

def add_wise_filters(df1, df2, columns=AstroTools.Filters['WISE']):
    
    df = df1.join(df2[columns])
    
    return df

def add_stellar_params(df1, df2, columns=stellar):
    
    df = df1.join(df2[columns])
    
    return df

def remove_contaminants(df):
    
    df1 = df.replace(np.nan, 'OK', regex=True)
    dfcp = df1[df1.subclass.str.contains('R|S|W') == False].copy()
    
    return dfcp

def remove_bad_values(df, Column = list(['SNR','FEH_ADOP','TEFF_ADOP','LOGG_ADOP', 
                                         'FEH_ADOP_UNC', 'TEFF_ADOP_UNC', 'LOGG_ADOP_UNC'])):
    
    df_clean = df[(df[Column[0]].gt(10)) &
                  (df[Column[1]].between(-5.0,1.0)) &
                  (df[Column[2]].between(4000,9200)) &
                  (df[Column[3]].between(0.0,5.0)) &
                  (df[Column[4]].between(0.0,0.25)) &
                  (df[Column[5]].between(0.0,200)) &
                  (df[Column[6]].between(0.0,0.3))].copy()
    
    return df_clean

def remove_spec_subclasses(df):
    
    #df['SPECTYPE_SUBCLASS'] = df.SPECTYPE_SUBCLASS.map(lambda x: x.rstrip(r'(\d)'))
    df = df.assign(SPECTYPE_SUBCLASS = df.subclass.str.replace(r'(\d)', '')) # Replace Original Column
    
    return df

def remove_bad_flags(df):
    
    dfcp = df[df.INSPECT.str.contains('nnnnnn') == True].copy()
    
    return dfcp
```

```python
df = AstroTools.Read_Dataset('JPLUS_DR2_nSSPP_LAMOST_DR5.csv')
df_result = pd.merge(df, acor_6, how='inner', on='TILE_ID')

Working_df = (df_result.pipe(correctMags)            # MW Correction
              .pipe(add_id_columns, df_result)       # Add TILE_ID, NUMBER, RA, DEC
              .pipe(order_columns)            # Reorder columns
              .pipe(add_gaia_filters, df_result)     # Add GAIA Filters              
              .pipe(add_wise_filters, df_result)     # Add WISE and 2MASS Filters
              .pipe(add_stellar_params, df_result, stellar)   # Add TEFF_ADOP, FEH_ADOP, LOGG_ADOP to Dataframe
              .pipe(remove_bad_values)        # Remove -9999.0 values
  #            .pipe(remove_bad_flags)
             )

Working_df.dropna(inplace=True)
Working_df.isnull().sum()
```

```python
from sklearn.base import TransformerMixin

class FeaturesSelection(TransformerMixin):   # Create custom transformer for features selection
    
    def __init__(self,cols):
        self.cols = cols
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        Xcols = X[self.cols]
        return Xcols
    
class MakeColors():                    # Create custom transformer for buid JPLUS colors
    def __init__(self, func):
        self.func = func

    def fit(self, X, y=None, **fit_params):
        return self
        
    def transform(self, X, **transform_params):
        return self.func(X)
    
class MakePristine():                    # Create custom transformer for buid JPLUS colors
    def __init__(self, func):
        self.func = func

    def fit(self, X, y=None, **fit_params):
        return self
        
    def transform(self, X, **transform_params):
        return self.func(X)
    
def Pristine(df):
    
    df = df.assign(J0395gSDSSgSDSSiSDSS = df.J0395 - df.gSDSS - 1.5*df.gSDSS + 1.5*df.iSDSS)
    
    return df
    
def Log10(X,column=0):

    X[:,column] = np.log10(X[:,column])
    
    return X

def Exp10(X,column=0):

    X[:,column] = exp10(X[:,column])
    
    return X
```
