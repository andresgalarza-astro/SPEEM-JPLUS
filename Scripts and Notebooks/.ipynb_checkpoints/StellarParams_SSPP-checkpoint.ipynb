```python
# Creating all the possible color combinations based on JPLUS Filters
from  itertools import combinations
Filters = columns[4:16]

df = DF[Filters].copy()

cc = list(combinations(df.columns,2))

df = pd.concat([df[c[0]].sub(df[c[1]]) for c in cc], axis=1, keys=cc)
df.columns = df.columns.map(''.join)
'''
df_narrow = Dataset[Narrowband_Filters]
df_broad = Dataset[Broadband_Filters]

cc_narrow = list(combinations(df_narrow.columns,2))
cc_broad = list(combinations(df_broad.columns,2))
df_narrow = pd.concat([df_narrow[c[0]].sub(df_narrow[c[1]]) for c in cc_narrow], axis=1, keys=cc_narrow)
df_broad = pd.concat([df_broad[c[0]].sub(df_broad[c[1]]) for c in cc_broad], axis=1, keys=cc_broad)
df_narrow.columns = df_narrow.columns.map(''.join)
df_broad.columns = df_broad.columns.map(''.join)
df = pd.concat([df_narrow,df_broad],axis=1)
df['Class'] = Dataset['CLASS']
'''
df['Teff'] = Dataset['TEFF_ADOP']
df['Logg'] = Dataset['LOGG_ADOP']
df['FeH'] = Dataset['FEH_ADOP']
print (df.head())
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error

def my_mae(n_estimators, max_leaf_nodes, max_features, criterion, train_X, val_X, train_y, val_y):
    model = RandomForestRegressor(n_estimators=n_estimators, max_leaf_nodes=max_leaf_nodes, max_features=max_features, criterion=criterion, random_state=1)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
from sklearn.model_selection import train_test_split

Train_Data, Test_Data = train_test_split(df)

train_X_Teff = Train_Data.drop(['Teff','Logg', 'FeH'],axis=1)
val_X_Teff = Test_Data.drop(['Teff','Logg', 'FeH'],axis=1)
train_y_Teff = Train_Data.Teff
val_y_Teff = Test_Data.Teff
```

```python
criterion = 'mae'
#n_estimators = 170
#max_features = None
max_leaf_nodes = None

Params = np.empty((0,3))

for n_estimators in [90, 100, 130, 140, 150, 160, 170, 180, 190, 200]:
    for max_features in [10, 20, 30, 40, 50, 60]:
        mae = my_mae(n_estimators, max_leaf_nodes, max_features, criterion, train_X_Teff, val_X_Teff, train_y_Teff, val_y_Teff)
        Iter_Params = np.array([[n_estimators,max_features,mae]])
        Params = np.append(Params,Iter_Params,axis=0)

minimo = min(Params[:,2])    
Best_Params = Params[Params[:,2] == minimo]
print("N estimators: %d \t Criterion: %s \t Max Features: %d \t MAE:  %.2f" %(Best_Params[0,0], criterion, Best_Params[0,1], Best_Params[0,2]))
```

```python
random_state = np.random.randint(0,100)

Teff_RF_Model = RandomForestRegressor(n_estimators=170, max_leaf_nodes=None, max_features=10, criterion='mae', random_state=random_state)
Teff_RF_Model.fit(train_X_Teff, train_y_Teff)
preds_val = Teff_RF_Model.predict(val_X_Teff)
mae = mean_absolute_error(val_y_Teff, preds_val)
```

```python
def AdjacentSubplots_Teff(X,Y,minX,maxX):
    
    x = np.arange(minX,maxX,50)
    x1 = np.arange(minX,maxX,50)
    y1 = np.zeros(len(x))
    
    fig, axs = plt.subplots(2, 1, sharex=True,gridspec_kw={'height_ratios': [3, 1]})
    # Remove horizontal space between axes
    fig.subplots_adjust(hspace=0)
    fig.set_figwidth(15)
    fig.set_figheight(15)

    # X = Dataset.Teff
    # Y = Dataset.TEFF_SPEC

    # Plot each graph, and manually set the y tick values
    axs[0].plot(X,Y,'ro')
    axs[0].plot(x,x,'b--')
    #axs[0].set_yticks(np.arange(-0.9, 1.0, 0.4))
    axs[0].grid(True)
    axs[0].set_xlim(minX,maxX)
    axs[0].set_ylim(minX,maxX)
    axs[0].set_ylabel('TEFF_ADOP',fontsize=18)
    axs[0].set_title('Random Forest Estimations VS TEFF Spectroscopic for SDSS',fontsize=22)

    axs[1].plot(X,Y-X,'ro')
    axs[1].plot(x1,y1,'b--')
    #axs[1].set_yticks(np.arange(0.1, 1.0, 0.2))
    axs[1].grid(True)
    axs[1].set_ylim(-1000, 1000)
    axs[1].set_xlabel('TEFF_RF',fontsize=18)
    axs[1].set_ylabel('DIFFERENCE',fontsize=18)

    plt.tight_layout()
    plt.show()
```
