```python
# Create JPLUS Colors Dataset
def colors(Dataset):
    # Creating all the possible color combinations based on JPLUS Filters
    from  itertools import combinations
    
    df = Dataset.copy()

    cc = list(combinations(df.columns,2))

    df = pd.concat([df[c[0]].sub(df[c[1]]) for c in cc], axis=1, keys=cc)
    df.columns = df.columns.map(''.join)
    
    return df
```

```python
Colors = Colors(DF[Filters])
```

```python
cor_logg = Colors.corrwith(DF.LOGG_ADOP)
SortedCorr = cor_logg.abs().sort_values(ascending=False, kind="mergesort").copy()

Selected_Features = SortedCorr.head(10).drop(SortedCorr.index[0]).index

top_corr_logg = SortedCorr.head(10).index

Colors['Log'] = DF['LOGG_ADOP'].copy()
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(Colors[np.append(top_corr_logg, ['Log'])].corr(),annot=True,cmap="RdYlGn")
```
