```python
Filters = list(AstroTools.Filters['JPLUS'])
Correction = list(AstroTools.Filters['Ax'])
A_Correction = list(acor_6.columns[1:])
    
dfcp = pd.concat([df_result[a].sub(df_result[b]).add(df_result[c]) for a, b, c in zip(Filters,Correction, A_Correction)],axis=1, keys=Filters).round(3)
```
